```python
cik_dict = {"SM":"0000893538", "FANG": "0001539838", "RSPP":"0001588216",
            "PE":"0001594466" , "MRO":"0000101778", "LPI":"0001528129", 
            "NFX":"0000912750", "GPOR":"0000874499", "RRC":"0000315852", "EQT":"0000033213 "}
all_ten_k = []
```

```python
def ten_k_docs(company,cik):
    company = edgar.Company(company, cik)
    tree = company.getAllFilings(filingType = "10-K")
    docs = edgar.getDocuments(tree, noOfDocuments=1)
    all_ten_k.append(docs)
```

```python
for key in cik_dict:
    comp = key
    cik = cik_dict[key]
    print(comp, cik)
    ten_k_docs(comp, cik)
```
